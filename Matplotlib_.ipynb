````python
import numpy as np

plt.figure(figsize=(7, 4))
X = np.linspace(-np.pi, np.pi, 1000)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, label='Cos')
plt.plot(X, S, label='Sin', c='r', linestyle='--', linewidth=4)
plt.xlabel('X', fontsize=16)
plt.ylabel('Y')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', 0, r'$\pi/2$', r'$\pi$'], fontsize=14)
plt.legend()
plt.show()
````

# Homework

The data in [populations.txt](http://www.scipy-lectures.org/_downloads/populations.txt) describes the populations of hares, lynxes and carrots in northern Canada during 20 years. Get the data with ```np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')``` and do the following exercises **without for-loops**.


Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.

````python
import matplotlib.pyplot as plt
import numpy as np
data=np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
plt.figure(figsize=(8, 4))
u, x, y, z=data[:, 0], data[:, 1], data[:, 2], data[:, 3]
plt.plot(u, x, label='hare',color='green', marker='o', markersize=5)
plt.plot(u, y, label='lynx', marker='o', markersize=5)
plt.plot(u, z, label='carrot',color='orange', marker='o', markersize=5)
plt.xticks(u[:], fontsize=7)
plt.xlabel('year')
plt.ylabel('population size ')
plt.legend()
plt.show()
````

Find which year each species had the largest population. Mark these years in the plot of populations.

````python
plt.figure(figsize=(8, 4))
u, x, y, z=data[:, 0], data[:, 1], data[:, 2], data[:, 3]
plt.plot(u, x, label='hare',color='green', marker='o', markersize=2)
plt.plot(u, y, label='lynx', marker='o', markersize=2)
plt.plot(u, z, label='carrot',color='orange', marker='o', markersize=2)
plt.xticks(u[:], fontsize=7)
for i in range(1, 4):
   plt.scatter(u[np.argmax(data[:, i])],data[np.argmax(data[:, i]), i],color='red',s=30)
plt.xlabel('year')
plt.ylabel('population size ')
plt.legend()
plt.show()
````

Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

````python
plt.figure(figsize=(8, 4))
u, x, y, z=data[:, 0], data[:, 1], data[:, 2], data[:, 3]
plt.xticks(u[:], fontsize=7)
plt.bar(u-0.2, x, color='green', width=0.4, label='hare', alpha=0.75)
plt.bar(u+0.2, y,  width=0.4, label='lynx', alpha=0.75)
plt.xlabel('year')
plt.ylabel('population size ')
plt.legend()
plt.show()
````

Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.

````python
u, x, y, z=data[:, 0], data[:, 1], data[:, 2], data[:, 3]
f, ax = plt.subplots(1, 3, sharey=True, figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.hist(x, 20, density=True, facecolor='pink', alpha=0.75, label='hare')
plt.xlabel('Values')
plt.ylabel('Probability Distribution')
plt.subplot(1, 3, 2)
plt.hist(y, 20, density=True, alpha=0.75, label='lynx')
plt.subplot(1, 3, 3)
plt.hist(y, 20, density=True, facecolor='orange',alpha=0.75, label='carrot')
for i in range(1, 4):
  plt.subplot(1, 3, i)
  plt.axvline(x=np.mean(data[:, i]), color='black')
  plt.legend()
  plt.grid(True)
print(f"std:{np.std(data[:, 1]):.3f} {np.std(data[:, 2]):.3f} {np.std(data[:, 3]):.3f}")
plt.show()
````

Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

````python
x, y=data[:, 1], data[:, 2]
h=0
l=0
mask = x>=y
new_list = list(map(lambda x: 'H'if x == True else 'L', mask))
print(new_list)
for i in range(20):
  if(mask[i]):
     h=h+1
value=[h, 20-h]
labels=['hare','lynx']
plt.pie(value, labels=labels, autopct='%1.1f%%')
plt.show()
````

Plot the change in the hare population and in the hare population for each year. Find the correlation coefficient between the both time series.

````python
plt.figure(figsize=(8, 4))
x, y= data[:, 1], data[:, 2]
plt.plot(u, x, label='hare',color='pink', marker='o', markersize=2)
plt.plot(u, y, label='lynx', color='orange',marker='o', markersize=2)
plt.xticks(u[:], fontsize=7)
plt.xlabel('year')
plt.ylabel('population size ')
plt.legend()
Corr=np.corrcoef(x, y)
print(f"correlation coefficient={Corr[0, 1]:.6f} ")
plt.show()
````

Using a scatter plot, show the population of hares vs lynxes by year (each point corresponds to a particular year, and the point coordinates correspond to the population of the two species in that year).

````python
plt.figure(figsize=(8, 4))
u, x, y=data[:, 0], data[:, 1], data[:, 2]
plt.xticks(u[:], fontsize=7)
sizes = np.random.uniform(80, 300, len(x))
plt.scatter(u , x, s=sizes, label='hare')
plt.scatter(u , y, s=sizes, c='orange', label='lynx')
plt.xlabel('year')
plt.ylabel('population size ')
plt.legend()
plt.show()
````

Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

````python
u, x=data[:, 0], data[:, 1]
pl=plt.hist(x[:-1], 20, density=True, facecolor='pink', alpha=0.75, label='hare')
sum=0
for i in range(20):
  sum+=pl[0][i]*(pl[1][i+1]-pl[1][i])
  if(sum > 0.5):
    break
value=pl[1][i]
print("ratio of the error to the true value=",abs(x[-1]-value)/x[-1])
plt.show()
````
